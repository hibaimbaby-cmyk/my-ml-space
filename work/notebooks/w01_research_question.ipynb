## 1. My lane (or freestyle) and why

*Name your lane — or say 'freestyle' and describe your own question. One short paragraph: why this one?*

```python
Lane: Content Decay / Refresh
Why: I chose this lane because maintaining existing organic traffic is often more cost-effective than acquiring new users from scratch. Over time, high-performing pages lose their rankings due to outdated information, algorithmic shifts, or competitor updates. By leveraging Machine Learning, we can proactively identify which declining pages have the highest potential for a successful refresh, helping the content team prioritize their optimization efforts where they will yield the maximum return on investment (ROI).
```

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
1. What decision does your work improve?
It improves the decision of selecting and prioritizing which older, declining web pages should be allocated budget and time for a content refresh, rather than guessing or relying purely on manual intuition.

2. Who acts on it?
The Content Marketing Team and SEO Specialists who are responsible for updating, rewriting, and optimizing the website's content.

3. What does a wrong recommendation cost?
A wrong recommendation (False Positive: predicting a page will recover when it won't) costs wasted editorial hours, content creation budget, and opportunity cost, as those resources could have been spent on pages with a higher likelihood of recovery. Alternatively, missing a good page (False Negative) means leaving potential organic traffic and revenue on the table.
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# قراءة البيانات مباشرة من الرابط الرسمي المعتمد للتدريب
url = "https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

# حساب الأرقام المطلوبة لمسار تحديث المحتوى (Content Decay)
old_pages = df[df["content_age_days"] > 365].shape[0]
avg_age = df["content_age_days"].mean()

print(f"Total pages older than 1 year: {old_pages}")
print(f"Average content age: {avg_age:.1f} days")
```
